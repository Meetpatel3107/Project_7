```python
import pandas as pd
```

```python
df = pd.read_csv('process_data.csv')
df.head()
```

```python
df['clean_text'] = df['clean_text'].fillna("")
```

```python
X = df['clean_text']
y = df['target']
```

```python
from sklearn.model_selection import  train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2 ,random_state=43)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
```

```python
vectorizer
```

```python
x_train_vz = vectorizer.fit_transform(X_train)
x_test_vz = vectorizer.fit_transform(X_test)
```

```python
x_train_vz,x_test_vz
```

```python
x_train_vz = vectorizer.fit_transform(X_train)
x_test_vz = vectorizer.transform(X_test)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation
from tensorflow.keras.optimizers import Adam
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_resample(x_train_vz, y_train)
```

```python
model = Sequential([
   Dense(128,activation='relu',input_shape=(x_train_vz.shape[1],)),
  Dropout(0.3),
   Dense(64,activation='relu'),
  Dropout(0.3),
  Dense(1,activation='sigmoid')
])
```

#comiple model

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
import tensorflow as tf
```

```python
X_train_resampled_dence = X_train_resampled.toarray() #convert to dense aary
```

```python
model.fit(X_train_resampled_dence,y_train_resampled,batch_size=32,epochs=10,validation_split=0.2)
```

```python
loss,accuracy = model.evaluate(x_test_vz,y_test)
print('Test Accuracy:',accuracy)
```
